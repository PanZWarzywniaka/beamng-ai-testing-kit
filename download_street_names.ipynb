```python
import json
```

```python
bbox = [53.291691, -1.808285, 53.551973, -1.216129]
```

```python
with open("sheffield_streets.txt") as f:
    lines = f.readlines() 

print(len(lines))
streets = {x.strip("\n") for x in lines if x!="\n"}
print(len(streets))
```

```python
json_dic = {
    "bbox": bbox,
    "streets": list(streets),
}
with open("streets.json", "w") as f:
    json.dump(json_dic, f)
json_dic
```
